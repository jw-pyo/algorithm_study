# Sorting Algorithm
- Problem: Sort a random array in decending order.

```python
import random
import time

ARR_SIZE = 4096

arr_rand = list(range(ARR_SIZE))
random.shuffle(arr_rand)

print(arr_rand)
```

## 1. Insertion Algorithm

- O(n^2)

```python
arr = arr_rand.copy()
timestamp = time.time()

for i in range(1, ARR_SIZE):
    val_cur = arr[i]
    j = i
    while j > 0:
        if val_cur > arr[j-1]:
            arr[j] = arr[j-1]
            j -= 1
        else:
            break
    idx_sorted = j
    arr[idx_sorted] = val_cur

arr_sorted = arr
print('Total time: ', time.time() - timestamp)
print('Sorted array', arr_sorted)
```

## 2. Merge sort
- Devide and conquer
- O(nlogn)

```python
INF = 987654321
idx_mid = lambda arr: len(arr) // 2

arr = arr_rand.copy()
timestamp = time.time()

def merge_sort(arr1, arr2):
    if len(arr1) > 1:
        arr1 = merge_sort(arr1[:idx_mid(arr1)], arr1[idx_mid(arr1):])

    if len(arr2) > 1:
        arr2 = merge_sort(arr2[:idx_mid(arr2)], arr2[idx_mid(arr2):])

    i = j = 0

    len_merged = len(arr1) + len(arr2)
    arr_new = [-1 for _ in range(len_merged)]
    
    arr1.append(-INF)
    arr2.append(-INF)

    for k in range(len_merged):
      
        if arr1[i] > arr2[j]:
            arr_new[k] = arr1[i]
            i = i+1
        else:
            arr_new[k] = arr2[j]
            j = j+1

    return arr_new

arr_sorted = merge_sort(arr[:idx_mid(arr)], arr[idx_mid(arr):])

print('Total time: ', time.time() - timestamp)
print('Sorted array', arr_sorted)
```

```python
arr = arr_rand.copy()
timestamp = time.time()

def exchange_element(arr, i, j):
    temp = arr[i]
    arr[i] = arr[j]
    arr[j] = temp

def min_heapify(arr, i, heap_size):
    child_l = 2*i
    child_r = 2*i + 1
    
    idx_low = i
    
    if child_l < heap_size and arr[child_l] < arr[i]:
        idx_low = child_l
        
    if child_r < heap_size and arr[child_r] < arr[idx_low]:
        idx_low = child_r
    
    if idx_low != i:
        exchange_element(arr, idx_low, i)
        min_heapify(arr, idx_low, heap_size)
    
    
def build_min_heap(arr):
    last_parent = len(arr) // 2
    
    for i in reversed(range(last_parent)):
        min_heapify(arr, i, len(arr))
        
def heap_sort(arr):
    build_min_heap(arr)
    heap_size = len(arr)
    for i in reversed(range(len(arr))):
        exchange_element(arr, 0, i)
        heap_size -= 1
        min_heapify(arr, 0, heap_size)
        
    return arr
        
arr_sorted = heap_sort(arr)
print('Total time: ', time.time() - timestamp)
print('Sorted array', arr_sorted)
```

## 4. Quicksort

- Worst case: O(n^2)
- Average case: O(nlogn)

```python
arr = arr_rand.copy()
timestamp = time.time()

def partition(arr, p, r):
    pivot = arr[r]
    i = p - 1
    
    for j in range(p, r):
        if arr[j] > pivot:
            i = i + 1
            exchange_element(arr, i, j)
            
    exchange_element(arr, i+1, r)
    q = i+1
    
    return q


def quicksort(arr, p, r):
    if p < r:
        pivot_rand = random.choice(list(range(p,r+1)))
        exchange_element(arr, pivot_rand, r)
        
        q = partition(arr, p, r)
        quicksort(arr, p, q-1)
        quicksort(arr, q+1, r)

    return arr
arr_sorted = quicksort(arr, 0, len(arr)-1)
print('Total time: ', time.time() - timestamp)
print('Sorted array', arr_sorted)
```
