# Sorting Algorithm
- Problem: Sort a random array in decending order.

```
import random
import time

ARR_SIZE = 4096

arr_rand = list(range(ARR_SIZE))
random.shuffle(arr_rand)

print(arr_rand)
```

## 1. Insertion Algorithm

- O(n^2)

```
arr = arr_rand.copy()
timestamp = time.time()

for i in range(1, ARR_SIZE):
    val_cur = arr[i]
    j = i
    while j > 0:
        if val_cur > arr[j-1]:
            arr[j] = arr[j-1]
            j -= 1
        else:
            break
    idx_sorted = j
    arr[idx_sorted] = val_cur

arr_sorted = arr
print('Total time: ', time.time() - timestamp)
print('Sorted array', arr_sorted)
```

# 2. Merge sort
- Devide and conquer
- O(nlogn)

```
INF = 987654321
idx_mid = lambda arr: len(arr) // 2

arr = arr_rand.copy()
timestamp = time.time()

def merge_sort(arr1, arr2):
    if len(arr1) > 1:
        arr1 = merge_sort(arr1[:idx_mid(arr1)], arr1[idx_mid(arr1):])

    if len(arr2) > 1:
        arr2 = merge_sort(arr2[:idx_mid(arr2)], arr2[idx_mid(arr2):])

    i = j = 0

    len_merged = len(arr1) + len(arr2)
    arr_new = [-1 for _ in range(len_merged)]
    
    arr1.append(-INF)
    arr2.append(-INF)

    for k in range(len_merged):
      
        if arr1[i] > arr2[j]:
            arr_new[k] = arr1[i]
            i = i+1
        else:
            arr_new[k] = arr2[j]
            j = j+1

    return arr_new

arr_sorted = merge_sort(arr[:idx_mid(arr)], arr[idx_mid(arr):])

print('Total time: ', time.time() - timestamp)
print('Sorted array', arr_sorted)
```
